```python
import pandas as pd 
import numpy as np
from sklearn.datasets import load_wine
```

```python
wine_data = load_wine()
```

```python
dir(wine_data)
```

```python
output=wine_data.target
```

```python
df=pd.DataFrame(inputs,columns=wine_data.feature_names)
```
